```python
import subprocess

# List of conflicting packages and their required versions
packages = [
    ("wurlitzer", ""),
    ("numpy", "==1.24.4"),
    ("dill", "==0.3.1.1"),
    ("pyarrow", "==9.0.0"),
    ("google-cloud-storage", "==2.2.1"),
    ("urllib3", "==1.26.15"),
    ("packaging", "==23.1"),
    ("shapely", "==2.0.1"),
    ("tensorflow", "==2.15.0"),
    ("ml-dtypes", "==0.3.1"),
    ("torch", "==2.1.2"),
]

# Install each package with the specified version
for package, version in packages:
    if version:
        subprocess.run(["pip", "install", f"{package}{version}"])
    else:
        subprocess.run(["pip", "install", package])
```
